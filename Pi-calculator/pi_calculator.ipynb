#### Let's load the packages first.

```python
import numpy as np
import random as rd
import math
```

#### Now, let's define some functions that we are going to use later.

```python
#The following function takes a point and calculates its
#distance from the center of the circle at (0.5, 0.5) 
def how_far_from_center(x, y):
    return math.sqrt((x-0.5)**2 + (y-0.5)**2)

#The following function generates a given number of pairs
#of random numbers between 0 and 1. Then checks which of
#those points lie within the cirle (r=0.5, c=(0.5, 0.5))
#pi = 4*(points in the circle / total points)
def find_pi_and_plot(num):
    total_points = num
    within_circle = 0 #This is a counter for points within the circle
    i=0
    while i<total_points:
        #Generate floating points within 0 and 1
        real = rd.random()
        imag = rd.random()
        radsq = how_far_from_center(real, imag)
    
        if radsq<0.5 :
            #In this case random point is within the circle.
            #Count this point, and plot it as a red dot.
            within_circle += 1
            ax.scatter(real, imag, s=1, color='red')
        else :
            #In this case random point is outside the circle.
            #Don't count this point, and plot it as a red dot.
            ax.scatter(real, imag, s=1, color='blue')
        i=i+1
    
    print(f'Numerator = {within_circle}')
    print(f'Denominator = {total_points}')
    frac = within_circle/total_points
    print(f'Fraction = {frac}')
    pi = frac*4
    print(f'Pi = {pi}')
    return pi
```

### Main part of the code:

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(5,5))

num = 1000
#This determines the accuracy of the method.
#With larger and larger numbers, it becomes more reliable.

pi = find_pi_and_plot(num)

#Let's also plot the circle for reference
circle = plt.Circle((0.5, 0.5), 0.5, color='black', fill=False)
ax.add_artist(circle)

#Some decoration=:
ax.set(xlim=(0, 1), xticks=np.arange(0, 1.1, 0.1),
       ylim=(0, 1), yticks=np.arange(0, 1.1, 0.1))
ax.set_title(f'$\pi$ = 4x(red points / total points ({num})) = {pi:.3f}')
plt.show()

print('done!')
```
